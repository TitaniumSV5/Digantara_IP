```python
#Preprocessing all the images in the dataset to create a new dataset

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

input_folder="/Users/anshumanmahapatra/Desktop/Digantara_IP/Datasets/Raw_Images"
output_folder="/Users/anshumanmahapatra/Desktop/Digantara_IP/Datasets/Preprocessed_Images"  # Save results here

alpha = 0.5  # Contrast
beta = 0.5   # Brightness
threshold1 = 100
threshold2 = 300

for filename in os.listdir(input_folder):
    if filename.endswith(".tiff"):
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)

        if img is None:
            print(f"Skipping unreadable file: {filename}")
            continue

        img_adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
        edges = cv2.Canny(img_adjusted, threshold1=threshold1, threshold2=threshold2)

        # Save edge image
        out_path = os.path.join(output_folder, f"edges_{filename.replace('.tiff', '.png')}")
        cv2.imwrite(out_path, edges)
        print(f"Processed and saved: {out_path}")
```
