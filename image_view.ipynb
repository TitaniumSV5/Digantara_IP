```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path="/Users/anshumanmahapatra/Desktop/Digantara_IP/Datasets/Raw_Images/Raw_Observation_001_Set1.tiff"
img=cv2.imread(img_path, cv2.IMREAD_UNCHANGED)

alpha=1.5
beta=50

img_adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

plt.figure(figsize=(10, 10))
plt.imshow(img_adjusted, cmap='gray')
plt.title("Image Display")
plt.axis('off')
plt.show()
```
